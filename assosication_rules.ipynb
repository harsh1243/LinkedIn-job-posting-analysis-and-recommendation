```python
keywords = [k.lower() for k in keywords]

def row_matches(skills):
    if not isinstance(skills, list):
        return False
    skills_lower = [s.lower() for s in skills]
    return any(k in skills_lower for k in keywords)

df_filtered = df[df["cleaned_skills"].apply(row_matches)]

print("Original rows:", len(df))
print("Filtered rows:", len(df_filtered))
df_filtered.head()
```

```python
# Simple FP-Growth on top-K skills + quick heatmap
from collections import Counter
import pandas as pd

# tune these
top_k = 200            # keep top 200 skills (lower if memory error)
min_support = 0.05     # e.g. itemset appears in >= 5% transactions
min_conf = 0.6         # rule confidence threshold

# 1. get transactions (list of lists)
transactions = df_filtered["cleaned_skills"].dropna().tolist()

# 2. keep only top_k most frequent skills
all_items = [it for tx in transactions for it in tx]
top_items = set([item for item, _ in Counter(all_items).most_common(top_k)])
tx_filtered = [[it for it in tx if it in top_items] for tx in transactions]
tx_filtered = [tx for tx in tx_filtered if tx]   # drop empty rows

print("Transactions:", len(transactions), "→ after filter:", len(tx_filtered), "unique items kept:", len(top_items))


from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
te_ary = te.fit(tx_filtered).transform(tx_filtered)
df_onehot = pd.DataFrame(te_ary, columns=te.columns_)

# 4. FP-Growth + rules
from mlxtend.frequent_patterns import fpgrowth, association_rules
freq_itemsets = fpgrowth(df_onehot, min_support=min_support, use_colnames=True)
rules = association_rules(freq_itemsets, metric="confidence", min_threshold=min_conf)

# 5. show top rules
cols = ["antecedents", "consequents", "support", "confidence", "lift"]
if rules.empty:
    print("No rules found — try lowering min_support or min_conf, or increase top_k.")
else:
    print("Top rules:")
    display(rules[cols].sort_values("confidence", ascending=False).head(20))

# 6. Quick heatmap for single-item -> single-item rules
import matplotlib.pyplot as plt
import seaborn as sns

single = rules[(rules["antecedents"].apply(len) == 1) & (rules["consequents"].apply(len) == 1)].copy()
if single.empty:
    print("No single→single rules for heatmap (try lowering thresholds).")
else:
    single["ant"] = single["antecedents"].apply(lambda s: list(s)[0])
    single["cons"] = single["consequents"].apply(lambda s: list(s)[0])
    heat = single.pivot_table(index="ant", columns="cons", values="confidence", fill_value=0)
    plt.figure(figsize=(8,6))
    sns.heatmap(heat, annot=True, fmt=".2f")
    plt.title("Confidence Heatmap (antecedent → consequent)")
    plt.tight_layout()
    plt.show()
```

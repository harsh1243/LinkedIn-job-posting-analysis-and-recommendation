```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("linkedin_job_postings.csv")


us_counts = (
    df[df['search_country'] == 'United States']['search_position']
    .value_counts()
    .head(10)
)


canada_counts = (
    df[df['search_country'] == 'Canada']['search_position']
    .value_counts()
    .head(10)
)


plt.figure(figsize=(12, 6))
us_counts.plot(kind='bar')
plt.title("Top 10 Job Titles in United States")
plt.xlabel("Job Title")
plt.ylabel("Count")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()


plt.figure(figsize=(12, 6))
canada_counts.plot(kind='bar', color='orange')
plt.title("Top 10 Job Titles in Canada")
plt.xlabel("Job Title")
plt.ylabel("Count")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
